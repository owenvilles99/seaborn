```python
import seaborn as sms
```

```python
import matplotlib.pyplot as plt
```

```python
matplotlib inline
```

```python
sms.set_style('darkgrid')
```

```python
titanic = sms.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sms.countplot(x= 'sex', data=titanic)
```

# boxplot

```python
sms.boxplot(x='class', y='age', data= titanic)
```

# swarmplot

```python
sms.swarmplot(x='class', y='age', data=titanic)
```

# matrix

```python
titanic = sms.load_dataset('titanic')
```

```python
titanic.head()
```

```python
tc = titanic.corr()
```

```python
sms.heatmap(tc)
```

```python
sms.heatmap(titanic.corr(), cmap='coolwarm')
```

# displot

```python
sms.displot(titanic['fare'])
```

# jointplot

```python
sms.jointplot(x='fare', y= 'age', data=titanic)
```

# displot

```python
sms.displot(titanic, x='age', row='sex',binwidth=5, height=5, facet_kws=dict(margin_titles=True),)
```

```python
g = sms.FacetGrid(data = titanic, col = 'sex')
g.map(plt.hist, 'age')
```

```python
g = sms.FacetGrid(data = titanic, col = 'sex')
g.map(sms.distplot, 'age')
```
